```python
#! lấy data đã được xử lý (lúc này phải chạy ipynb trong datapreprocessing) cho vào all_movie_link:
all_movie_link = []
with open('../DataPreprocessing/Processed_movie_link.txt','r',encoding="utf8") as file:
    all_movie_link = file.readlines()
```

```python
#! vào từng đường link một để lấy toàn bộ thông tin về phim đó
for url in all_movie_link:
    

    page = requests.get(url)

    soup = BeautifulSoup(page.text,'html')
    path = soup.find_all('div',attrs={'class':'spaceit_pad'})
    my_in4 = []
    for in4 in path:
        # in4 = [item.strip() for item in in4 if str(item)]
        a = in4.get_text().replace('\n','')
        #print(a)
        my_in4.append([a])
```

```python

#! lưu in4 vào dạng csv
import csv
with open('in4_movie.csv','w',encoding='utf-8') as file:
    
    for i in my_in4:
        csvwriter = csv.writer(file)
        csvwriter.writerow(i)
```
